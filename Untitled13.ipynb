```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.dpi"] = 120

print("Done")
```

```python
SENTIMENT_FILE = os.path.join(FOLDER, "fear_greed_index.csv")
TRADES_FILE    = os.path.join(FOLDER, "historical_data.csv")

sentiment = pd.read_csv(SENTIMENT_FILE)
trades    = pd.read_csv(TRADES_FILE)

print(f"Sentiment rows : {len(sentiment):,}  |  columns: {list(sentiment.columns)}")
print(f"Trades rows    : {len(trades):,}  |  columns: {list(trades.columns)}")
```

```python
print("─── Sentiment (first 5 rows) ───")
display(sentiment.head())

print("\n─── Trades (first 5 rows) ───")
display(trades.head())
```

```python

def find_col(df, candidates):
    """Return the first matching column (case-insensitive, stripped)."""
    normalized = {c.strip().lower(): c for c in df.columns}
    for name in candidates:
        if name.strip().lower() in normalized:
            return normalized[name.strip().lower()]
    return None

# Strip whitespace from all column names
sentiment.columns = sentiment.columns.str.strip()
trades.columns    = trades.columns.str.strip()

# Drop duplicate columns — keep first occurrence
dupes = [col for col, count in Counter(trades.columns).items() if count > 1]
if dupes:
    print(f"Dropping duplicate columns: {dupes}")
trades    = trades.loc[:, ~trades.columns.duplicated(keep="first")]
sentiment = sentiment.loc[:, ~sentiment.columns.duplicated(keep="first")]

# ── Sentiment date ────────────────────────────────────────────
sent_date_col = find_col(sentiment, ["Date", "date", "DATE", "Datetime", "timestamp"])
if sent_date_col is None:
    raise KeyError(f"No date column found in sentiment.\nColumns: {list(sentiment.columns)}")

sentiment = sentiment.rename(columns={sent_date_col: "Date"})
sentiment["Date"] = pd.to_datetime(sentiment["Date"]).dt.date

# ── Trades timestamp & PnL ────────────────────────────────────
trade_time_col = find_col(trades, ["Timestamp IST", "Timestamp", "time", "datetime", "Date"])
trade_pnl_col  = find_col(trades, ["Closed PnL", "closedPnL", "closed_pnl", "pnl", "PnL"])

if trade_time_col is None:
    raise KeyError(f"No time column found.\nColumns: {list(trades.columns)}")
if trade_pnl_col is None:
    raise KeyError(f"No PnL column found.\nColumns: {list(trades.columns)}")

trades = trades.rename(columns={trade_time_col: "time", trade_pnl_col: "closedPnL"})

trades["time"]      = pd.to_datetime(trades["time"].astype(str), format="%Y-%m-%d %H:%M:%S")
trades["Date"]      = trades["time"].dt.date
trades["closedPnL"] = pd.to_numeric(trades["closedPnL"], errors="coerce")

null_pnl = trades["closedPnL"].isna().sum()
if null_pnl > 0:
    print(f"Warning: {null_pnl} rows had non-numeric PnL → set to NaN.")

print("Cleaning done ✓")
print(f"\nSample parsed timestamps:")
print(trades[["time", "Date", "closedPnL"]].head(5).to_string(index=False))
```

```python

daily_pnl = trades.groupby("Date")["closedPnL"].sum().reset_index()
merged    = daily_pnl.merge(sentiment, on="Date", how="inner")

print(f"Merged: {len(merged)} days  |  {merged['Date'].min()} → {merged['Date'].max()}")
display(merged.head(10))
```

```python
print(merged.columns.tolist())
print(merged.head(3))
```

```python

avg_pnl = (
    merged.groupby("classification")["closedPnL"]
    .mean()
    .sort_values(ascending=False)
)

print("Average daily PnL by sentiment:\n")
print(avg_pnl.to_string())
```

```python

fig, ax = plt.subplots(figsize=(10, 5))
avg_pnl.plot(kind="bar", ax=ax, color="steelblue", edgecolor="white")

ax.set_title("Average Daily PnL by Market Sentiment", fontsize=14, pad=12)
ax.set_xlabel("Sentiment", labelpad=8)
ax.set_ylabel("Average PnL (USD)")
ax.tick_params(axis="x", rotation=30)
ax.axhline(0, color="black", linewidth=0.8, linestyle="--")

plt.tight_layout()
plt.savefig(os.path.join(FOLDER, "avg_pnl_by_sentiment.png"))
plt.show()
print("Saved → avg_pnl_by_sentiment.png")
```

```python
order = avg_pnl.index.tolist()

fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=merged, x="classification", y="closedPnL", order=order, ax=ax)

ax.set_title("PnL Distribution by Sentiment", fontsize=14, pad=12)
ax.set_xlabel("Sentiment", labelpad=8)
ax.set_ylabel("Daily PnL (USD)")
ax.tick_params(axis="x", rotation=30)
ax.axhline(0, color="red", linewidth=0.8, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.savefig(os.path.join(FOLDER, "pnl_distribution.png"))
plt.show()
print("Saved → pnl_distribution.png")
```

```python
volatility = (
    merged.groupby("classification")["closedPnL"]
    .std()
    .sort_values(ascending=False)
)

print("PnL volatility (std dev) by sentiment:\n")
print(volatility.to_string())
```

```python

best     = avg_pnl.idxmax()
worst    = avg_pnl.idxmin()
most_vol = volatility.idxmax()

print("── Summary ──────────────────────────────────")
print(f"Best avg PnL   : {best:20s}  ${avg_pnl[best]:>10,.2f}")
print(f"Worst avg PnL  : {worst:20s}  ${avg_pnl[worst]:>10,.2f}")
print(f"Most volatile  : {most_vol:20s}  σ = ${volatility[most_vol]:>8,.2f}")
print(f"Total days     : {len(merged)}")
print(f"Date range     : {merged['Date'].min()}  →  {merged['Date'].max()}")
```
